```python
import numpy as np
from specdb.specdb import SpecDB, IgmSpec
from astropy.io import fits

from funcs_query import query, pmf, identify
```

```python
igmsp = SpecDB('/media/bartosz/USB STICK/BOSS_DR14/IGMspec_DB_v03.1.hdf5')
meta = igmsp['BOSS_DR14'].meta
meta[1]
```

```python
fpath = '/media/bartosz/USB STICK/dr16q_prop_May16_2023.fits'

# Open the .fits file
with fits.open(fpath) as hdul:
    # Print information about the file
    hdul.info()

    # Access the data from the primary HDU (header/data unit)
    data1 = hdul[1].data

    # Optionally, you can also access the header
    header1 = hdul[1].header
    print(header1)
```

```python
meta_data = np.loadtxt('/media/bartosz/USB STICK/BOSS_DR14_ext/meta_data_ext_combined.txt',dtype='str')
idx = [int(i) for i in meta_data[:,0]]
Z = [float(z) for z in np.array(meta_data)[:,1]]
```

```python
j=0
new_meta=[]
for k,i in enumerate(idx):
    j+=1
    if Z[k] < 1.8:
        continue
    idx_dr16 = query(meta,data1,i)
    if idx_dr16 == None:
        continue
    lbol = data1['LOGLBOL'][idx_dr16]
    if lbol > 46.6355:
        meta_row = np.append(meta_data[k],lbol)
        new_meta.append(meta_row)
        print(j)
    else: continue
```

```python
with open('/media/bartosz/USB STICK/BOSS_DR14_ext/meta_data_lbol_10%.txt', 'w') as save:
    for x in np.array(new_meta):
        save.write('{} \t {} \t {} \t {} \t {} \t {} \t {} \n'.format(*x))
save.close()
```
